```python
import torch
from torch import nn, optim
```

```python
random_input = torch.randn(3,3)
random_output = torch.randn(3,3)
criterion = nn.MSELoss()
# model
l1 = nn.Linear(3, 10, bias=False)
l2 = nn.Linear(10, 3, bias=False)
intermediate = l1(random_input)
# intermediate.requires_grad=False  RuntimeError
intermediate = intermediate.detach()
output = l2(intermediate)
print(l1)
loss = criterion(output ,random_output)
loss.backward()
```

```python
random_input = torch.randn(1)
random_output = torch.randn(1)

class AXB(nn.Module):
    def __init__(self):
        super(AXB, self).__init__()
        self.a = nn.Parameter(torch.ones(1))
        self.b = nn.Parameter(torch.ones(1))
    
    def forward(self, x):
        return self.a * x + self.b
    
class CXD(nn.Module):
    def __init__(self):
        super(CXD, self).__init__()
        self.c = nn.Parameter(torch.ones(1))
        self.d = nn.Parameter(torch.ones(1))
        self.axb = AXB()
    
    def forward(self, x):
        return self.c * self.axb(x) + self.d

cxd = CXD()
cxd.axb.b.requires_grad = False
optimizer = optim.SGD([{'params': cxd.c}, {'params': cxd.d}, {'params': cxd.axb.parameters()}], lr=0.01, momentum=0.9)    
loss = nn.MSELoss()
print(cxd.c.data)
print(cxd.d.data)
print(cxd.axb.a.data)
print(cxd.axb.b.data)
```

```python
out = cxd(torch.ones(1))

output = loss(out, torch.ones(1) * 100)
optimizer.zero_grad()
output.backward()
optimizer.step()
```
